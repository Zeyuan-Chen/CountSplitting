## Compare PMF of Poisson and Negative binomial with different level of over dispersion
poisson and 3 negative binomial distribution: <br>

$Poisson(\Lambda = 5)$<br>
$Negative Binomial(\Lambda = 5, b = 0.5)$<br>
$Negative Binomial(\Lambda = 5, b = 5)$<br>
$Negative Binomial(\Lambda = 5, b = 50)$<br>

in dnbinom:  $\mu = \Lambda$ $\text{size} = b$

```r
options(scipen = 999, digits = 2) # sig digits

n = 0:15
df <- data.frame(events = 0:15, 
                 NB1      = dnbinom(x = n, mu = 5, size = 0.5),
                 NB2      = dnbinom(x = n, mu = 5, size = 5  ),
                 NB3      = dnbinom(x = n, mu = 5, size = 50 ),
                 Poisson  = dpois  (x = n, lambda = 5))   
```

```r
df
```

```r
df_tidy <- gather(df, key = "Distribution", value = "density", -c(events))
df_tidy$Distribution = factor(df_tidy$Distribution, levels = c("NB1", "NB2", "NB3", "Poisson"))

g = ggplot(df_tidy, aes(x = factor(events), y = density, fill = Distribution)) +
    geom_col(position = "dodge") +
    
    theme_classic() + 
    #ggtitle(expression(paste("Negative Binomial under different overdispersion:  ", over(Lambda, b)))) + 
    ggtitle(expression(paste("Negative Binomial under different overdispersion"))) + 
    theme(plot.title = element_text(hjust = 0.5)) + 
    labs(x = "Events (k)", y = "P(x = k)") + 

    scale_fill_discrete(guide = guide_legend(title = ""), 
                        labels = unname(TeX(c("$NB(\\Lambda = 5, \\; b = 0.5$)", 
                                              "$NB(\\Lambda = 5, \\; b = 5 )$", 
                                              "$NB(\\Lambda = 5, \\; b = 50 $)",
                                              "$Poisson(\\Lambda = 5)$")))) + 
    theme(legend.position = c(.95, .8), 
          legend.justification = c("right", "top"),
          legend.text.align = 0)
```

```r
options(repr.plot.width = 5, repr.plot.height = 4, repr.plot.res = 250)
g
```
